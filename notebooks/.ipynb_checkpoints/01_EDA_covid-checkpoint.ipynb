```python
#Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10,6)

DATA = Path("../data")
OUT = Path("../outputs")
OUT.mkdir(exist_ok=True)
```

```python
#load raw CSVs
confirmed = pd.read_csv(DATA / "time_series_covid19_confirmed_global.csv")
deaths = pd.read_csv(DATA / "time_series_covid19_deaths_global.csv")
print("confirmed:", confirmed.shape)
print("deaths:", deaths.shape)
confirmed.head()
```

```python
#inspect column layout
confirmed.columns[:10]
```

```python
#melt confirmed (wide â†’ long)
id_cols = ['Province/State','Country/Region','Lat','Long']
date_cols = [c for c in confirmed.columns if c not in id_cols]

conf_long = confirmed.melt(id_vars=id_cols, value_vars=date_cols,
                           var_name='date', value_name='confirmed')
conf_long['date'] = pd.to_datetime(conf_long['date'], format='%m/%d/%y', errors='coerce')
conf_long.head()
```

```python
#melt deaths
date_cols_d = [c for c in deaths.columns if c not in id_cols]
death_long = deaths.melt(id_vars=id_cols, value_vars=date_cols_d,
                         var_name='date', value_name='deaths')
death_long['date'] = pd.to_datetime(death_long['date'], format='%m/%d/%y', errors='coerce')
death_long.head()
```

```python
#merge confirmed + deaths
df = pd.merge(conf_long, death_long,
              on=['Province/State','Country/Region','Lat','Long','date'],
              how='outer')

df = df.rename(columns={
    'Province/State': 'province',
    'Country/Region': 'country',
    'Lat': 'lat',
    'Long': 'long'
})
df = df.sort_values(['country','province','date']).reset_index(drop=True)
df.head()
```

```python
#aggregate to country level
country_df = df.groupby(['country','date'], as_index=False).agg({
    'confirmed': 'sum',
    'deaths': 'sum'
}).sort_values(['country','date'])
country_df.head()
```

```python
#compute daily new cases/deaths
country_df['new_confirmed'] = country_df.groupby('country')['confirmed'].diff().fillna(0)
country_df['new_deaths'] = country_df.groupby('country')['deaths'].diff().fillna(0)

# handle negative diffs (reporting corrections) â€” set to 0 for simplicity
country_df['new_confirmed'] = country_df['new_confirmed'].clip(lower=0)
country_df['new_deaths'] = country_df['new_deaths'].clip(lower=0)

country_df.tail()
```

```python
#quick summary & save cleaned CSV
print("Rows, cols:", country_df.shape)
country_df.to_csv(DATA / "jhu_covid_country_daily.csv", index=False)
print("Saved cleaned CSV to:", DATA / "jhu_covid_country_daily.csv")
```

```python
#sample plot: daily new cases for a few countries
sample_countries = ['India','US','Brazil','United Kingdom','Germany']
sample = country_df[country_df['country'].isin(sample_countries)].copy()

# ensure date is datetime
sample['date'] = pd.to_datetime(sample['date'])

# 7-day rolling average
sample['new_confirmed_7d'] = sample.groupby('country')['new_confirmed'].transform(lambda s: s.rolling(7, min_periods=1).mean())

sns.lineplot(data=sample, x='date', y='new_confirmed_7d', hue='country')
plt.title("7-day avg new confirmed (sample countries)")
plt.xticks(rotation=30)
plt.tight_layout()
plt.savefig(OUT / "new_cases_7d_sample_countries.png")
plt.show()
```

```python
#Top 10 countries by total confirmed cases and deaths
latest = country_df[country_df['date'] == country_df['date'].max()]

top_cases = latest.nlargest(10, 'confirmed')
top_deaths = latest.nlargest(10, 'deaths')

fig, axes = plt.subplots(1, 2, figsize=(15, 6))
sns.barplot(data=top_cases, x='confirmed', y='country', ax=axes[0])
axes[0].set_title('Top 10 Countries by Total Confirmed Cases')

sns.barplot(data=top_deaths, x='deaths', y='country', ax=axes[1])
axes[1].set_title('Top 10 Countries by Total Deaths')

plt.tight_layout()
plt.savefig(OUT / "top10_cases_deaths.png")
plt.show()
```

```python
#Case Fatality Rate (CFR)
latest['CFR'] = (latest['deaths'] / latest['confirmed']) * 100
cfr_top = latest[latest['confirmed'] > 100000].nlargest(10, 'CFR')

sns.barplot(data=cfr_top, x='CFR', y='country', palette='Reds_r')
plt.title("Countries with Highest Case Fatality Rate (CFR)")
plt.xlabel("CFR (%)")
plt.savefig(OUT / "cfr_top_countries.png")
plt.show()
```

```python
#Trends for India
country = "India"
india = country_df[country_df['country'] == country]
india['new_confirmed_7d'] = india['new_confirmed'].rolling(7).mean()
india['new_deaths_7d'] = india['new_deaths'].rolling(7).mean()

plt.figure(figsize=(10,5))
sns.lineplot(x='date', y='new_confirmed_7d', data=india, label='New Cases (7d avg)')
sns.lineplot(x='date', y='new_deaths_7d', data=india, label='New Deaths (7d avg)')
plt.title(f"{country}: Daily New COVID-19 Cases and Deaths (7-day Avg)")
plt.xlabel("Date")
plt.ylabel("Count")
plt.legend()
plt.savefig(OUT / f"{country.lower()}_trend.png")
plt.show()
```

```python
#Correlation Heatmap
corr = country_df[['confirmed','deaths','new_confirmed','new_deaths']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation between Key COVID-19 Metrics")
plt.savefig(OUT / "correlation_heatmap.png")
plt.show()
```
